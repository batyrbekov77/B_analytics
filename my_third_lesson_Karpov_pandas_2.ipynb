### Метод .apply()

```python
user_df = user_df \
    .rename(columns = {
        'tc':'user_ID',
        'art_sp':'brand_info'
    })
user_df
```

```python
brand_name = 'MARAVILLA 500 G Store_Brand'
brand_name.split(' ')[-1]
```

```python
def split_brand(brand_name_data):
    return brand_name_data.split(' ')[-1]
```

```python
split_brand('MARAVILLA 500 G Store_Brand')
```

```python
user_df['brand_name'] = \
    user_df \
        .brand_info \
        .apply(split_brand)
```

```python
user_df.head()
```

### Функция лямбда

```python
# более компактный способ получить имя бренда

user_df['brand_name'] = \
    user_df \
        .brand_info \
        .apply(lambda x: x.split(' ')[-1])
```

```python
user_df
```

### Приступаем к аналитике
#### Исходная задача: мы хотим проверить гипотезу о покупках, где есть пользователи, которые являются лояльными пользователями к некоторому брэнду.

```python
users_purchases = \
    user_df \
        .groupby('user_ID', as_index = False) \
        .agg({'brand_name':'count'}) \
        .rename(columns = {'brand_name':'purchases'}) \
        .query('purchases >= 5')
```

```python
users_purchases.head()
```

```python
users_unique_brands = \
    user_df \
        .groupby('user_ID', as_index = False) \
        .agg({'brand_name':pd.Series.nunique}) \
        .rename(columns = {'brand_name':'unique_brand'})
```

```python
users_unique_brands.head()
```

```python
lovely_brand_purchases = \
    user_df \
        .groupby(['user_ID', 'brand_name'], as_index = False) \
        .agg({'brand_info':'count'}) \
        .sort_values(['user_ID', 'brand_info'], ascending = [False, False]) \
        .groupby('user_ID') \
        .head(1) \
        .rename(columns = {'brand_name':'lovely_brand', 'brand_info':'lovely_brand_purchases'})
```

```python
lovely_brand_purchases
```

### Задание 2:
#### Какой клиент совершил больше всего успешных операций? (success == True)

```python
logs
```

```python
success_logs = \
    logs \
        .query('success == True') \
        .groupby('client', as_index = False) \
        .agg({'success':'count'}) \
        .sort_values('success', ascending = False) \
        .rename(columns = {'client': 'id', 'success': 'count_success'}) \
        .query('count_success == 41')
success_logs
```

```python
mmm = success_logs.count_success.max()
success_logs['bbb'] = success_logs['count_success'] >= mmm
success_logs
```

```python
logs_client = logs[['client', 'success']]
logs_client.query('success == True').value_counts()
```
