### Задача 2:
#### Преобразование в нижний регистр и в нижний подчеркивания

```python
# создаем небольшую фукнцию для
#    1. преобразование в нижниый регистр: .lower()
#    2. замена пробела в нижниый подчеркивания: .rename()

def replace_space_with_star(name):
    new_name = name \
                    .lower() \
                    .replace(' ', '_')
    return new_name 
booking = booking.rename(columns = replace_space_with_star)
booking
```

### Задача 3:
#### Топ-5 стран по успешным бронированиям

```python
# выводим топ-5 стран по успешным бронированиям

booking_success = booking \
    .query('is_canceled == 0') \
    .groupby('country', as_index = False) \
    .agg({'is_canceled':'count'}) \
    .sort_values('is_canceled', ascending = False) \
    .head(5)
```

### Задача 8:
#### Создайте колонку total_kids, объединив children и babies. Для отелей какого типа среднее значение переменной оказалось наибольшим? В качестве ответа укажите наибольшее среднее total_kids, округлив до 2 знаков после точки

```python
booking['total_kids'] = booking['children'] + booking['babies']
max_hotel_by_children = booking \
        .groupby('hotel', as_index = False) \
        .agg({'total_kids':'mean'}) \
        .round(2) \
        .max()
max_hotel_by_children
```
